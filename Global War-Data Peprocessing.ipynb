### One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

### Ordinal Encoding

```python
#Climate_Zone
from sklearn.preprocessing import OrdinalEncoder
cl = [['Tropical','Arid','Temperate','Polar']]
od = OrdinalEncoder(categories=cl)
od.fit(df[['Climate_Zone']])
df['or_Climate_Zone']= od.transform(df[['Climate_Zone']])
```

```python
df['Weapons_Used'].unique()
```

```python
wu =[['Cyber','Mixed','Conventional','Nuclear']]
odw = OrdinalEncoder(categories=wu)
df['or_Weapon_Used']=odw.fit_transform(df[['Weapons_Used']])
```

```python
df['Terrain_Type'].unique()
```

```python
tr ={'Plains':1,'Desert':2,'Urban':3,'Forest':4,'Mountain':5}
df['or_Terrain_Type']= df['Terrain_Type'].map(tr)
```

```python
df.drop(columns=['Terrain_Type','Climate_Zone','Weapons_Used'], inplace=True)
```

```python
#IQR Method
q1 = df['Longitude'].quantile(0.25)
q3 = df['Longitude'].quantile(0.75)
Iqr = q3 - q1
min_range = q1 - (1.5 * Iqr)
max_range = q3 + (1.5 * Iqr)
print(min_range,max_range)
```
